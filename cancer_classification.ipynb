```python
X=df.drop('benign_0__mal_1',axis=1).values
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
# 30 features hain
```

```python
model=Sequential()
# binary hai jbhi end layer m sigmoid add krre
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
model.history.history
```

```python
losses=pd.DataFrame(model.history.history)
losses
```

```python
losses.plot()
# overfitting
```

# early stopping

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python

early_stop=EarlyStopping(monitor='val_loss',mode='min',
                        verbose=1,patience=25)
```

```python
model=Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
losses=model.history.history
```

```python
loss=pd.DataFrame(losses)
loss
```

```python
loss.plot()
```

# dropout layers
ye layers o turn off krdega

```python
model=Sequential()
# binary hai jbhi end layer m sigmoid add krre
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
loss=pd.DataFrame(model.history.history)
loss
```

```python
loss.plot()
# this is even better itna flat krdiya its good 
```

```python
# Get the predicted probabilities (between 0 and 1)
predictions = model.predict(X_test)

# Convert probabilities to binary class labels (0 or 1)
predicted_classes = (predictions > 0.5).astype(int)

print(predicted_classes.flatten())  # Flatten to print class labels like [1 1 1 0 0 1 1]
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predicted_classes))
print('\n')
print(classification_report(y_test,predicted_classes))
```
